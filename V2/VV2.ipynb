```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def realS(x): return x*np.exp(-x)
```

```python
def F(x, y, dy): return -dy-2*y-np.exp(-x)
```

```python
def mEU(a, b, N, alph):
    NODs = np.linspace(a, b, N+1)
    VALsY, VALsV = np.empty(N+1), np.empty(N+1)
    h, VALsY[0], VALsV[0] = (b-a)/N, realS(a), alph
    for i in range(N):
        VALsY[i+1] = VALsY[i] + h*VALsV[i]
        VALsV[i+1] = VALsV[i] + h*F(NODs[i], VALsY[i], VALsV[i])
    return NODs, VALsY
```

```python
def SMwBis(a, b, N, eps, nmeth):
    alph1, alph2, k, yb = 0.1, -0.1, 1, realS(b)
    while (nmeth(a, b, N, alph1)[1][-1] - yb)*(nmeth(a, b, N, alph2)[1][-1] - yb) > 0:
        k+=1
        alph1 *= k
        alph2 *= k
    if alph1 > alph2: alph1, alph2 = alph2, alph1
    NODs, VALs = nmeth(a, b, N, alph1)
    F1 = VALs[-1] - yb
    if abs(F1) <= eps: return NODs, VALs
    NODs, VALs = nmeth(a, b, N, alph2)
    F2 = VALs[-1] - yb
    if abs(F2) <= eps: return NODs, VALs
    while abs(VALs[-1] - yb) > eps:
        alph = 0.5*(alph1 + alph2)
        NODs, VALs = nmeth(a, b, N, alph)
        if F1*(VALs[-1] - yb) < 0: alph2, F2 = alph, VALs[-1] - yb
        else: alph1, F1 = alph, VALs[-1] - yb
    return NODs, VALs
```

```python
def Test1(a, b, N, eps, nmeth):
    rX = np.linspace(a, b, 100)
    rY = realS(rX)
    nX, nY = SMwBis(a, b, N, eps, nmeth)
    plt.plot(rX, rY, label = "$y(x)$")
    plt.plot(nX, nY, label = "$SMwBis$")
    plt.grid(True)
    plt.axis([a, b, min(rY)-1, max(rY)+1])
    plt.legend()
    plt.show()
```

```python
Test1(-1, 5, 20, 0.0001, mEU)
```

```python
rX = np.linspace(-1, 5, 100)
nX, nY = mEU(-1, 5, 100, 2*2.7182818)
plt.plot(rX, realS(rX), label = "$y(x)$")
plt.plot(nX, nY, label = "$mEU$")
plt.grid(True)
plt.legend()
plt.show()
```
